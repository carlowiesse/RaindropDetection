## Preprocessing
### Decide which preprocessing technique to use
**First step:** The paper about this network doesn't specify which technique the authors used for preprocessing. So I decided to compare three of the best approaches according to a research study.

**Reference:** K. K. Pal and K. S. Sudeep, *Preprocessing for image classification by convolutional neural networks,* IEEE International Conference on Recent Trends in Electronics, Information & Communication Technology, 2016.

```python
# load all the training images as a 4D array (required for tensorflow)
X = load_data(train_img_paths,size=128)
```

```python
X_pixel_norm = X/255. # Pixel normalization
X_mean_norm = X - X.mean(axis=0) # Mean normalization
X_std_norm = X_mean_norm/np.maximum(0.001,X.std(axis=0)) # Standardization

idx = 2
for i in [idx,idx+503]:
    
    fig, ax = plt.subplots(1,4,figsize=(16,4))
    plt.setp(ax,xticks=[],yticks=[],frame_on=False)

    ax[0].imshow(X[i],cmap='gray')
    ax[0].set_title('No Normalization')

    ax[1].imshow(X_pixel_norm[i],cmap='gray')
    ax[1].set_title('Pixel Normalization')

    ax[2].imshow(X_mean_norm[i],cmap='gray')
    ax[2].set_title('Mean Normalization')
    
    ax[3].imshow(X_std_norm[i],cmap='gray')
    ax[3].set_title('Standardization')

    fig.tight_layout()
    plt.show()
```

**Observation:** The two images on the left look the same, and the two on the right look almost the same. So I learned that I need to plot these images as 3D plots to see the effects of the different normalization techniques.

### Plot normalized images as 3D plots to better differenciate these techniques
**Second step:** Plot the image as a 3D plot by using the image coordinates as x and y values and the pixel value as the z value. With this, I will decide which techniques are more beneficial for differentiating between clear and raindrop images.

```python
X_pixel_norm = X/255.
X_mean_norm = X - X.mean(axis=0)
X_std_norm = X_mean_norm/np.maximum(0.001,X.std(axis=0))

idx = 2
for i in [idx,idx+503]:

    fig = plt.figure(figsize=(32,8))
    plot_parameters = {'s':5,'depthshade':False}

    ax1 = fig.add_subplot(141, projection='3d')
    x,y,z = array2xyz(X[i])
    ax1.scatter(x,y,z,c=z,cmap='gray',**plot_parameters)
    ax1.set_title('No Normalization')

    ax2 = fig.add_subplot(142, projection='3d')
    x,y,z = array2xyz(X_pixel_norm[i])
    ax2.scatter(x,y,z,c=z,cmap='gray',**plot_parameters)
    ax2.set_title('Pixel Normalization')

    ax3 = fig.add_subplot(143, projection='3d')
    x,y,z = array2xyz(X_mean_norm[i])
    ax3.scatter(x,y,z,c=z,cmap='gray',**plot_parameters)
    ax3.set_title('Mean Normalization')
    
    ax3 = fig.add_subplot(144, projection='3d')
    x,y,z = array2xyz(X_std_norm[i])
    ax3.scatter(x,y,z,c=z,cmap='gray',**plot_parameters)
    ax3.set_title('Standardization')

    plt.show()
```

**Observation:** Based on the plots above, I learned that pixel normalization and standardization are better options when trying to difference between clear and raindrop images. On the other hand, mean normalization seems to make it harder to differentiate between the two.

## Implementing CNN model from paper
**Third step:** Recreate the neural network model from the paper.

```python
seed = 420
np.random.seed(seed) # Set random seed for replicability

# load tensorflow libraries
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

import tensorflow as tf
tf.keras.backend.clear_session()  # For easy reset of notebook state
tf.get_logger().setLevel('ERROR') # Remove Tensorflow unnecessary warnings
```

```python
def build_model():
    
    '''
    Model proposed in the research paper for raindrop detection
    '''

    xin = Input(shape=(128,128,1))

    # Block 1
    x = Conv2D(64, kernel_size=(3, 3),activation='relu',padding='same')(xin)
    x = Conv2D(64, kernel_size=(3, 3),activation='relu',padding='same')(x)
    x = MaxPooling2D((2, 2),padding='same')(x)
    
    # Block 2
    x = Conv2D(32, kernel_size=(3, 3),activation='relu',padding='same')(x)
    x = Conv2D(32, kernel_size=(3, 3),activation='relu',padding='same')(x)
    x = MaxPooling2D((2, 2),padding='same')(x)
    
    # Block 3
    x = Conv2D(16, kernel_size=(3, 3),activation='relu',padding='same')(x)
    x = Conv2D(16, kernel_size=(3, 3),activation='relu',padding='same')(x)
    x = Conv2D(16, kernel_size=(3, 3),activation='relu',padding='same')(x)
    x = MaxPooling2D((2, 2),padding='same')(x)
    
    # Block 4
    x = Conv2D(8, kernel_size=(3, 3),activation='relu',padding='same')(x)
    x = Conv2D(8, kernel_size=(3, 3),activation='relu',padding='same')(x)
    x = Conv2D(8, kernel_size=(3, 3),activation='relu',padding='same')(x)
    x = AveragePooling2D((2, 2),padding='same')(x)
    
    x = Flatten()(x) # 4D -> 2D array (required for "Dense" layer)
    
    # Fully-connected layers
    x = Dense(110, activation='relu')(x)
    x = Dense(50, activation='relu')(x)
    x = Dense(2, activation='softmax')(x)
    
    model = Model(xin,x,name='my_model')
    return model
```

## Training the model
**Fourth step:** Train the neural network using our training data. I train 5 instances of proposed model, each with a different fifth of the data used as validation to avoid data bias during training.

```python
# Apply standardization
X_norm = (X - X.mean(axis=0))/np.maximum(0.001,X.std(axis=0))
# load true labels of the training data 
y = np.array([0 if l==labels[0] else 1 for l in train_img_labels]) 
```

```python
# reduce learning rate of model after 5 unsuccessful training steps
reduce_lr = ReduceLROnPlateau(monitor='val_loss',patience=5, verbose=1, min_lr=1e-6)
# stop training after 7 unsuccessful training steps
early_stop = EarlyStopping(monitor='val_loss',patience=7, verbose=1, min_delta=1e-3)
```

```python
# split data into 5 groups, 4 for training and 1 validation
skf = StratifiedKFold(n_splits=5)

for i,(train, validation) in enumerate(skf.split(X_norm, y)):
    print('\n-- Model %s --' % i)
    # compile model
    model = build_model()
    model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
    # train model
    history = model.fit(X_norm[train],to_categorical(y)[train],
                        validation_data=(X_norm[validation],to_categorical(y)[validation]),
                        epochs=100,
                        batch_size=16,
                        callbacks=[reduce_lr,early_stop])
    # save model instance
    model.save_weights('split_'+str(i)+'_weights.h5')
```

## Test classification on validation data
**Fifth step:** Test the classification capabilities of each model instance using their corresponding validation data.

```python
fig, ax = plt.subplots(1,5,figsize=(20,3))
plt.setp(ax,xticks=[],yticks=[],frame_on=False)
colors = ['Blues','Reds','Purples','Greens','Greys']

# Apply standardization
X_norm = (X - X.mean(axis=0))/np.maximum(0.001,X.std(axis=0))
# load true labels of the training data 
y = np.array([0 if l==labels[0] else 1 for l in train_img_labels])

# split data into 5 groups, 4 for training and 1 validation
skf = StratifiedKFold(n_splits=5)
for i,(train, validation) in enumerate(skf.split(X_norm, y)):
    # load trained model
    model = build_model()
    model.load_weights('../models/cnn_models/split_%d_weights.h5' % i)
    
    # classify between clear or raindrop images
    y_pred = model.predict(X_norm[validation])
    y_pred_labels = [labels[np.argmax(i)] for i in y_pred]
    
    # plot confusion matrix
    cm = confusion_matrix(np.array(train_img_labels)[validation],y_pred_labels,labels=labels)
    sns.heatmap(cm, annot=True,xticklabels=['Clear','Rain'], yticklabels=['Clear','Rain'], cmap=colors[i], ax=ax[i])
    ax[i].set_title('CNN Model %d' % i)
    
plt.show()
```

**Observation:** The best instances are 0 and 4, as both only failed once. Therefore, I learned that I can use model instance 0 or 4 for the final evaluation.

## Define proposed pipeline
**Sixth step:** Create the final pipeline so that an RGB image is taken as input and the output is a predicted label. For the pipeline below, I used model instance 1 to classify between clear or raindrop images.

```python
def pipeline(query_image):
    
    im_b = query_image[...,2] # get blue channel

    # gamma correction
    gamma = 0.2
    im_gamma = im_b/255.
    im_gamma = np.power(im_gamma,gamma)
    im_gamma = np.clip(255*im_gamma,0,255)
    im_gamma = normalize(im_gamma)*255
    im_gamma = im_gamma.astype(np.uint8)
    
    im = resize_with_border(im_gamma,size=128) # resize image
    
    # load normalization factors used for standardization
    X_mean = np.load('../models/cnn_models/mean_norm_factor.npy')
    X_std = np.load('../models/cnn_models/std_norm_factor.npy')
    
    X = im.reshape(1,im.shape[0],im.shape[1],1) # reshape image into 4D array
    X_norm = (X - X_mean)/np.maximum(0.001,X_std) # standardization
    
    # load trained model
    model = build_model()
    model.load_weights('../models/cnn_models/split_1_weights.h5')

    # classify between clear or raindrop images
    prediction = model.predict(X_norm)
    predicted_label = 'LABELLED_CLEAR_INPUTS' if np.argmax(prediction[0])==0 else 'LABELLED_REAL_RAIN_INPUTS'
    
    return predicted_label
```

## Failed cases
**Eighth step:** Check the images that were wrongly classified to see what caused the mistake.

```python
y_pred = [pipeline(cv2.cvtColor(cv2.imread(i),cv2.COLOR_BGR2RGB)) for i in test_img_paths]
failed_cases = [i for i in range(len(test_img_paths)) if test_img_labels[i]!=y_pred[i]]
print('Image index of failed predictions: ',failed_cases)

fig, ax = plt.subplots(1,2,figsize=(12,6))
plt.setp(ax,xticks=[],yticks=[],frame_on=False)

for ix,i in enumerate(failed_cases):
    im = cv2.imread(test_img_paths[i]) # load image as BGR
    im_b = im[...,0] # get blue channel

    # gamma correction
    gamma = 0.2
    im_gamma = im_b/255.
    im_gamma = np.power(im_gamma,gamma)
    im_gamma = np.clip(255*im_gamma,0,255)
    im_gamma = normalize(im_gamma)*255
    im_gamma = im_gamma.astype(np.uint8)

    im = resize_with_border(im_gamma,128) # resize image

    ax[ix].imshow(im,cmap='gray')
    ax[ix].set_title('Index %d - Label %s' % (i,y_pred[i][9:-7]))

fig.tight_layout()
plt.show()
```

**Observation:** 

- For the left image, (1) the light going beetween the leaves resembled the light refraction caused by raindrops and (2) the overall blur of the image due to high speeds recreated the blur often found inside raindrops. This caused the image to be perceived as raindrops.

- For the right image, (1) the texture inside the raindrops merged with the trees and road as they all shared similar pixel values, causing the image to be perceived as clear.
